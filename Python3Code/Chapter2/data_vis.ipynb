```python
import pandas as pd
import os
```

```python
current_dir = os.path.abspath('')
data_dir = os.path.abspath(os.path.join(current_dir, '..'))
data_dir = os.path.abspath(os.path.join(data_dir, '..', 'Data/Martin_straight_14_leg 2024-06-07 14-42-51/Accelerometer.csv'))

#data_dir = os.path.abspath(os.path.join(os.path.dirname(__file__), '..', 'Data/Martin_straight_14_leg 2024-06-07 14-42-51'))

#print(data_dir)
df = pd.read_csv(data_dir)
```

```python
print(df.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Convert 'time' column to datetime and set as index
df['Time (s)'] = pd.to_datetime(df['Time (s)'])
# df = filtered_df.set_index('time')

# Plot the 'value' column
df['X (m/s^2)'].plot(figsize=(10, 6), title=f'X-Acceleration')


plt.xlabel('Time')
plt.ylabel('Accel')
plt.show()

df['Y (m/s^2)'].plot(figsize=(10, 6), title=f'Y-Acceleration')


plt.xlabel('Time')
plt.ylabel('Accel')
plt.show()

df['Z (m/s^2)'].plot(figsize=(10, 6), title=f'Z-Acceleration')


plt.xlabel('Time')
plt.ylabel('Accel')
plt.show()

df.boxplot()
```
